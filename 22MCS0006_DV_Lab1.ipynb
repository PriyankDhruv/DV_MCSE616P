```python
from matplotlib import pyplot as plt
import numpy as np

course = np.array(['Python', 'ML', 'COA', 'AI', 'DL', 'DSA'])
enrollment = np.array([30, 65, 23, 78, 32, 45])

plt.bar(course, enrollment)
#plt.scatter(course, enrollment, label = "SET-1")
plt.title('First Graph')
plt.xlabel('X - axis')
plt.ylabel('Y - axis')

plt.show()
```

```python
data = dict()

data['COA'] = 84
data['DSA'] = 100
data['OS'] = 88
data['DBA'] = 96
data['DL'] = 68

subjects = list(data.keys())
marks = list(data.values())

plt.bar(range(0, len(data)), marks, tick_label = subjects)
plt.show()
```

```python
s1 = [98, 68, 92, 78, 100, 68]
s2 = [96, 92, 100, 56, 99, 88]
s3 = [94, 98, 70, 100, 84, 99]

plt.plot(course, s1, label = 'Raju', linestyle = 'dotted')
plt.plot(course, s2, label = 'Kalu', linestyle = 'dashed')
plt.plot(course, s3, label = 'Bapu', linestyle = 'dashdot')

plt.title('Student Performance')
plt.xlabel('Subject')
plt.ylabel('Marks')

plt.legend()
plt.show()
```

```python
c1 = np.array([98, 68, 92, 78, 100, 68])
c2 = np.array([96, 92, 100, 56, 99, 88])
c3 = np.array([94, 98, 70, 100, 84, 99])

plt.scatter(course, s1, label = 'Raju', s = c1)
plt.scatter(course, s2, label = 'Kalu', s = c2)
plt.scatter(course, s3, label = 'Bapu', s = c3)

plt.title('Student Performance')
plt.xlabel('Subject')
plt.ylabel('Marks')

plt.legend()
plt.show()
```
