Anscombeâ€™s quartet comprises four datasets that have nearly identical simple statistical properties, yet appear very different when graphed. It was constructed in 1973 by the statistician Francis Anscombe to demonstrate both the importance of graphing data before analyzing it and the effect of outliers on statistical properties.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#populate the data (x and y)
x1=np.array([10,8,13,9,11,14,6,4,12,7,5])
x4=np.array([8,8,8,8,8,8,8,19,8,8,8])
y1=np.array([8.04,6.95,7.58,8.81,8.33,9.96,7.24,4.26,10.84,4.82,5.68])
y2=np.array([9.14,8.14,8.74,8.77,9.26,8.1,6.13,3.1,9.13,7.26,4.74])
y3=np.array([7.46,6.77,12.74,7.11,7.81,8.84,6.08,5.39,8.15,6.42,5.73])
y4=np.array([6.58,5.76,7.71,8.84,8.47,7.04,5.25,12.5,5.56,7.91,6.89])
```

```python
#plot data set-1
plt.scatter(x1,y1,c="blue")
plt.show()

# if you look at the scatter plot you will see that there seems to be a linear relationship between x and y
# Observation: Fits the linear regression model
```

```python
plt.scatter(x1,y2,c="green")
plt.show()
# if you look at this figure you can conclude that there is a non-linear relationship between x and y
# Observation: can't fit linear regression model because data show non-linear
```

```python
plt.scatter(x1,y3,c="orange")
plt.show()
# when there is a perfect linear relationship for all the data points except one which seems to be an outlier 
# which is indicated be far away from that line
# Observation: Outliers involved, can't handled by linear regression model
```

```python
plt.scatter(x4,y4,c="red")
plt.show()
# when one high-leverage point is enough to produce a high correlation coefficient
# Observation: Outliers involved, can't handled by linear regression model
```

```python
# Statistical information of the given data set: (mean, standard deviation and pearson correlation)

# Calculating the mean and standard deviation for all four data set
array =np.array([[np.mean(x1),np.std(x1),np.mean(y1),np.std(y1)],
                [np.mean(x1),np.std(x1),np.mean(y2),np.std(y2)],
                [np.mean(x1),np.std(x1),np.mean(y3),np.std(y3)],
                [np.mean(x4),np.std(x4),np.mean(y4),np.std(y4)]])
              
print(array)
```

```python
# Calculating the pearson correlation for all four data set
pcc1 = np.corrcoef(x1,y1)
pcc2 = np.corrcoef(x1,y1)
pcc3 = np.corrcoef(x1,y1)
pcc4 = np.corrcoef(x1,y1)
print(pcc1,pcc2,pcc3,pcc4)
```
